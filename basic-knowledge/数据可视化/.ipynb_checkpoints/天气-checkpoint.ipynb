```python
import csv
import matplotlib.pyplot as plt

filename = 'data/sitka_weather_07-2018_simple.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)#the first line of the file
    
    #get the highest temperatures in file
    highs= []
    for row in reader:#for each row, add 5th element to highs, namely the TMAX
        high = int(row[5])#turn into int
        highs.append(high)

#根据最高温度绘制图形
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.plot(highs, c='red')

# set the format of the fig
ax.set_title("Daily maximum temperature in July 2018",fontsize=24)
ax.set_xlabel('',fontsize =16)
ax.set_ylabel('temperature(F)',fontsize=16)
ax.tick_params(axis='both', which='major',labelsize=16)

plt.show()
```

在图表中添加日期

```python
import csv
from datetime import datetime
import matplotlib.pyplot as plt

filename = 'data/sitka_weather_07-2018_simple.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)#the first line of the file
    
    #get the highest temperatures in file
    dates, highs = [], []
    for row in reader:#for each row, add 5th element to highs, namely the TMAX
        current_date = datetime.strptime(row[2], '%Y-%m-%d')
        high = int(row[5])#turn into int
        dates.append(current_date)
        highs.append(high)

#根据最高温度绘制图形
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.plot(dates, highs, c='red')

# set the format of the fig
ax.set_title("Daily maximum temperature in July 2018",fontsize=24)
ax.set_xlabel('',fontsize =16)
fig.autofmt_xdate()#绘制倾斜的日期标签避免重叠
ax.set_ylabel('temperature(F)',fontsize=16)
ax.tick_params(axis='both', which='major',labelsize=16)

plt.show()
```

再绘制一个数据TLOW

```python
import csv
from datetime import datetime
import matplotlib.pyplot as plt

filename = 'data/sitka_weather_2018_simple.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)#the first line of the file
    
    #get the highest temperatures in file
    dates, highs, lows = [], [], []
    for row in reader:#for each row, add 5th element to highs, namely the TMAX
        current_date = datetime.strptime(row[2], '%Y-%m-%d')
        high = int(row[5])#turn into int
        low = int(row[6])
        dates.append(current_date)
        highs.append(high)
        lows.append(low)

#根据最高温度和最低温度绘制图形
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.plot(dates, highs, c='red')
ax.plot(dates, lows, c='blue')

# set the format of the fig
ax.set_title("Daily maximum temperature in 2018",fontsize=24)
ax.set_xlabel('',fontsize =16)
fig.autofmt_xdate()#绘制倾斜的日期标签避免重叠
ax.set_ylabel('temperature(F)',fontsize=16)
ax.tick_params(axis='both', which='major',labelsize=16)

plt.show()
```

图表区域着色

```python
import csv
from datetime import datetime
import matplotlib.pyplot as plt

filename = 'data/sitka_weather_2018_simple.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)#the first line of the file
    
    #get the highest temperatures in file
    dates, highs, lows = [], [], []
    for row in reader:#for each row, add 5th element to highs, namely the TMAX
        current_date = datetime.strptime(row[2], '%Y-%m-%d')
        high = int(row[5])#turn into int
        low = int(row[6])
        dates.append(current_date)
        highs.append(high)
        lows.append(low)

#根据最高温度和最低温度绘制图形
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.plot(dates, highs, c='red', alpha=0.5)#alpha指颜色的透明度
ax.plot(dates, lows, c='blue', alpha=0.5)
ax.fill_between(dates, highs, lows, facecolor='blue', alpha=0.1)

# set the format of the fig
ax.set_title("Daily maximum temperature in 2018",fontsize=24)
ax.set_xlabel('',fontsize =16)
fig.autofmt_xdate()#绘制倾斜的日期标签避免重叠
ax.set_ylabel('temperature(F)',fontsize=16)
ax.tick_params(axis='both', which='major',labelsize=16)

plt.show()
```

错误检查

```python
import csv
from datetime import datetime
import matplotlib.pyplot as plt

filename = 'data/death_valley_2018_simple.csv'#该数据集有一天有缺失，无法处理其最高温度
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)#the first line of the file
    
    #get the highest temperatures in file
    dates, highs, lows = [], [], []
    for row in reader:#for each row, add 5th element to highs, namely the TMAX
        current_date = datetime.strptime(row[2], '%Y-%m-%d')
        #处理缺失数据
        try:
            high = int(row[4])#turn into int
            low = int(row[5])
        except ValueError:
            print(f"Missing data for{current_date}")
        else:
            dates.append(current_date)
            highs.append(high)
            lows.append(low)

#根据最高温度和最低温度绘制图形
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.plot(dates, highs, c='red', alpha=0.5)#alpha指颜色的透明度
ax.plot(dates, lows, c='blue', alpha=0.5)
ax.fill_between(dates, highs, lows, facecolor='blue', alpha=0.1)

# set the format of the fig
ax.set_title("Daily maximum and minimum temperatures in 2018\nDeath Valley, California, USA",fontsize=24)
ax.set_xlabel('',fontsize =16)
fig.autofmt_xdate()#绘制倾斜的日期标签避免重叠
ax.set_ylabel('temperature(F)',fontsize=16)
ax.tick_params(axis='both', which='major',labelsize=16)

plt.show()
```
