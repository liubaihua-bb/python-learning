```python
import matplotlib.pyplot as plt
```

```python
squares = [1,4,9,16,25]
fig, ax = plt.subplots()
# subplots 可以在一张图片中绘制一个或者多个图表
# fig表示整张图片，ax便是图片中的各个图表
ax.plot(squares)
# plot尝试根据给定的数据以有意义的方式绘制图表

plt.show()
```

```python
squares = [1,4,9,16,25]

fig, ax = plt.subplots()

ax.plot(squares, linewidth=3)#linewidth：绘制线条粗细
# 设置图表标题，并给坐标轴加上标签
ax.set_title("Square number", fontsize=24)
ax.set_xlabel("value", fontsize=14)
ax.set_ylabel("Square of values", fontsize=14)
# 设置刻度标记的大小，axes影响x轴和y轴的刻度，labelsize刻度标记字号
ax.tick_params(axis='both', labelsize=14)
plt.show()
```

```python
input_values = [1,2,3,4,5]
squares = [1,4,9,16,25]

fig, ax = plt.subplots()
ax.plot(input_values, squares, linewidth=3)#linewidth：绘制线条粗细 #确定输入输出值矫正图像

#设置图表标题并加坐标轴标签
ax.set_title("Square number", fontsize=24)
ax.set_xlabel("value", fontsize=14)
ax.set_ylabel("Square of values", fontsize=14)

# 设置刻度标记的大小
ax.tick_params(axis='both', labelsize=14)
plt.show()
```

```python
input_values = [1,2,3,4,5]
squares = [1,4,9,16,25]

plt.style.use('seaborn')#内置样式seaborn

fig, ax = plt.subplots()
ax.plot(input_values, squares, linewidth=3)#linewidth：绘制线条粗细 #确定输入输出值矫正图像

#设置图表标题并加坐标轴标签
ax.set_title("Square number", fontsize=24)
ax.set_xlabel("value", fontsize=14)
ax.set_ylabel("Square of values", fontsize=14)

# 设置刻度标记的大小
ax.tick_params(axis='both', labelsize=14)
plt.show()
```

scatter散点图

```python
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(2, 4)#传递一对x和y坐标，绘制单个点

plt.show()
```

```python
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(2, 4, s=200)#s设置绘制图形时使用的点的尺寸

#设置图表标题并加坐标轴标签
ax.set_title("Square number", fontsize=24)
ax.set_xlabel("value", fontsize=14)
ax.set_ylabel("Square of values", fontsize=14)

# 设置刻度标记的大小
ax.tick_params(axis='both', which='major', labelsize=14)

plt.show()
```

```python
x_values = [1,2,3,4,5]
y_values = [1,4,9,16,25]

plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=100)

#设置图表标题并加坐标轴标签
ax.set_title("Square number", fontsize=24)
ax.set_xlabel("value", fontsize=14)
ax.set_ylabel("Square of values", fontsize=14)

# 设置刻度标记的大小
ax.tick_params(axis='both', which='major', labelsize=14)

plt.show()
```

```python
# 绘制1000个点
x_values = range(1,1001)
y_values = [x ** 2 for x in x_values]

plt.style.use('seaborn')
fig, ax = plt.subplots()
# ax.scatter(x_values, y_values, c='red', s=10)#c:点的颜色
# ax.scatter(x_values, y_values, s=10)
ax.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Blues, s=10)#c设置为y值列表，并使用参数cmap告诉pyplot使用哪个颜色映射


#设置图表标题并加坐标轴标签
ax.set_title("Square number", fontsize=24)
ax.set_xlabel("value", fontsize=14)
ax.set_ylabel("Square of values", fontsize=14)

# 设置刻度标记的大小
ax.tick_params(axis='both', which='major', labelsize=14)

# plt.show()
plt.savefig('squares_plot.png', bbox_inches='tight')#bbox_inches='tight'裁掉多余空白部分
```
